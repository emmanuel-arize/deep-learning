```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 3, 0.1)
y=x**3 -1/x
x
```

```python
with plt.style.context('ggplot'):
    plt.plot(x,y,'r-o',label='x**3 -1/x')
    plt.plot(x,3*x**2+1/x**-2)
    plt.plot( 1,'b')
plt.show()
```

```python
plt.rcParams['savefig.facecolor'] = "1"
def example_plot(ax, fontsize=12):
    ax.plot([])
    ax.plot(x,y)
    ax.locator_params(nbins=3)
    ax.set_xlabel('x-label', fontsize=fontsize)
    ax.set_ylabel('y-label', fontsize=fontsize)
    ax.set_title('Title', fontsize=fontsize)
plt.close('all')
fig, ax = plt.subplots()
example_plot(ax, fontsize=24)
```

```python
plt.plot(x,x**3 -1/x)
plt.plot([100])
plt.text(6,300,r"$x^{3} -\frac{1}{x}$")
plt.grid()
plt.legend()
plt.show()
```
