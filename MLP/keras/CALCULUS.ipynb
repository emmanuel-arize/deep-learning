# Tangent Lines
	
A tangent line to the function f(x) at the point x = a is a line that just touches the graph of the function at the point in question and is “parallel” (in some way) to the graph at that point


Consider the problem of trying to find an equation of the tangent line to a curve with
equation $y=f(x)$ at a given point $p$ .
Since we know that the point lies on the tangent line, we can find the equation of if we
know its slope .  we need two points to compute the slope but we know only one point $p$ on $t$. To get around the problem we first find an approximation to $m$ by taking a nearby point $Q$ on the curve and compute the slope $m_{pQ}$ of the secant line connecting $P$ and $Q$
$$M_{PQ}=\frac{f(x)-f(a)}{x-a}=f(a) + f'(a)(x-a) $$
 we would like an estimate that is at least
somewhat close the actual value so to get a better estimate we can take move $Q$ closer and closer to $P$.

 In this case slope of the secant line connecting $Q$ and
$P$ should be getting closer and closer to the slope of the tangent line and we write as
$$M=lim_{Q \rightarrow P } M_{PQ} $$

we say $M$ is the limit of $M_{PQ}$ as $Q$ approaches $P$ along the curve.
$$ M=lim_{x \rightarrow a}\frac{f(x) - f(a)}{x-a} $$

taking values of x that get closer and closer to x = a  on both
sides of x = a and use this list of values to estimate the slope of the tangent line, m.
The tangent line will then be,
$$ y =f(a) + m(x-a )$$

# instantaneous rate of change
under rate of change we want to determine how fast f(x) is changing at some point, say x = a 

in order to compute the average rate of change of $f(x)$ at $x = a$ we do this by choosing another point, say x, and then the average rate of change will be

$$ AVG-RATE-CHANGE=\frac{change \ in \ f(x)}{change \ in \ x}  $$
$$ AVG-RATE-CHANGE=\frac{f(x) -f(a)}{x-a}  $$

$$ M=lim_{x \rightarrow a}\frac{f(x) - f(a)}{x-a} $$

# Derivatives and Differentiation
IN DEEP LEARNING we typically choose loss functions that are differentiable with respect to our modelʼs parameters.

The dervivate of the function $f(x)$ with respect to $x$ is the function $f^{'}(x)$ defined as

$$f^{'}(x)=lim _{h\rightarrow 0} \frac{f(x+h)-f(x)}{h} $$

 Find the derivative of the following function using the definition of the derivative.
$$f(x)= 2x-  35$$

using the definition
$$f^{'}(x)=lim _{h\rightarrow 0} \frac{2(x+h)-35-(2x-35)}{h} $$
expanding the equation yields
$$f^{'}(x)=lim _{h\rightarrow 0} \frac{2x+2h-35-2x-35}{h} $$
$$f^{'}(x)=lim _{h\rightarrow 0} \frac{2h}{h} $$

$$f^{'}(x)=lim _{h\rightarrow 0} 2 =2$$


# DEFINITION
A function $f ( x)$  is called differentiable at $x = a$ if $f ′( x)$ exists and $f ( x)$ is called
differentiable on an interval if the derivative exists for each point in that interval

# Alternate Notation FOR DERIVATES
$$f^{'}(x)=y' =\frac{dy}{dx}=\frac{d}{dx}(y)=\frac{d}{dx}(f(x)) $$

If $f(x)= x^{n}$  then  $f′(x)= nx^{n−1}$ and is sometimes called the power rule

# Product Rule
If the two functions f(x) and g(x) are differentiable (i.e. the derivative exist) then the product is
differentiable and,
$$( fg)'= f'g + g'f$$


$$ (f g h)'= f'gh + fg'h + fgh'$$
$$(f g h w)'= f' g h w + f g' h w + f g h' w + f g h w'$$

# Quotient Rule
If the two functions f(x) and g(x) are differentiable (i.e. the derivative exist) then the quotient is
differentiable and,
$$ (\frac{f}{g})' =\frac{f'g-g'f}{g^{2}} $$

As a toy example, say that we are interested in differentiating the function $y = 3x^{T}x$ with respect
to the column vector x.

```python
# initializing x
x=tf.range(4)
x=tf.cast(x,dtype=tf.float32)
a=2*np.dot(x,x)
a,2*tf.tensordot(x,x,axes=1)
```

```python
with tf.GradientTape() as g:
    # trace the tensor x
    g.watch(x)
    y=3*tf.tensordot(x,x,axes=1)
# compute the derivative of y with respect to the x.
dy_dx=g.gradient(y,x)
dy_dx
```
